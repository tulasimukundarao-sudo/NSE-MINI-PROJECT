```python
# -------------------------------
# Expense Tracker System
# Python Mini Project
# -------------------------------

expenses = [
    {"ID": 1, "Category": "Food", "Amount": 250, "Date": "2025-01-01"},
    {"ID": 2, "Category": "Transport", "Amount": 120, "Date": "2025-01-02"},
    {"ID": 3, "Category": "Books", "Amount": 500, "Date": "2025-01-03"},
    {"ID": 4, "Category": "Food", "Amount": 300, "Date": "2025-01-04"},
    {"ID": 5, "Category": "Entertainment", "Amount": 450, "Date": "2025-01-05"},
    {"ID": 6, "Category": "Transport", "Amount": 200, "Date": "2025-01-06"},
    {"ID": 7, "Category": "Food", "Amount": 150, "Date": "2025-01-07"},
    {"ID": 8, "Category": "Stationery", "Amount": 220, "Date": "2025-01-08"},
    {"ID": 9, "Category": "Food", "Amount": 280, "Date": "2025-01-09"},
    {"ID": 10, "Category": "Bills", "Amount": 900, "Date": "2025-01-10"},
    {"ID": 11, "Category": "Transport", "Amount": 180, "Date": "2025-01-11"},
    {"ID": 12, "Category": "Food", "Amount": 260, "Date": "2025-01-12"},
    {"ID": 13, "Category": "Entertainment", "Amount": 600, "Date": "2025-01-13"},
    {"ID": 14, "Category": "Books", "Amount": 750, "Date": "2025-01-14"},
    {"ID": 15, "Category": "Food", "Amount": 320, "Date": "2025-01-15"},
    {"ID": 16, "Category": "Transport", "Amount": 140, "Date": "2025-01-16"},
    {"ID": 17, "Category": "Bills", "Amount": 1100, "Date": "2025-01-17"},
    {"ID": 18, "Category": "Food", "Amount": 200, "Date": "2025-01-18"},
    {"ID": 19, "Category": "Stationery", "Amount": 180, "Date": "2025-01-19"},
    {"ID": 20, "Category": "Food", "Amount": 270, "Date": "2025-01-20"},
    {"ID": 21, "Category": "Transport", "Amount": 160, "Date": "2025-01-21"},
    {"ID": 22, "Category": "Entertainment", "Amount": 550, "Date": "2025-01-22"},
    {"ID": 23, "Category": "Food", "Amount": 290, "Date": "2025-01-23"},
    {"ID": 24, "Category": "Books", "Amount": 680, "Date": "2025-01-24"},
    {"ID": 25, "Category": "Bills", "Amount": 950, "Date": "2025-01-25"},
    {"ID": 26, "Category": "Food", "Amount": 310, "Date": "2025-01-26"},
    {"ID": 27, "Category": "Transport", "Amount": 190, "Date": "2025-01-27"},
    {"ID": 28, "Category": "Stationery", "Amount": 210, "Date": "2025-01-28"},
    {"ID": 29, "Category": "Food", "Amount": 260, "Date": "2025-01-29"},
    {"ID": 30, "Category": "Entertainment", "Amount": 700, "Date": "2025-01-30"}
]

# Function to add an expense
def add_expense():
    try:
        expense_id = int(input("Enter Expense ID: "))
        category = input("Enter Category: ").strip().title()
        amount = float(input("Enter Amount: "))
        date = input("Enter Date (YYYY-MM-DD): ").strip()

        expense = {
            "ID": expense_id,
            "Category": category,
            "Amount": amount,
            "Date": date
        }

        expenses.append(expense)
        print("Expense added successfully!")

    except ValueError:
        print("Invalid input! ID and Amount must be numbers.")

# Function to view all expenses
def view_expenses():
    if len(expenses) == 0:
        print("No expense records found.")
    else:
        print("\n--- Expense Records ---")
        for expense in expenses:
            print("ID:", expense["ID"])
            print("Category:", expense["Category"])
            print("Amount:", expense["Amount"])
            print("Date:", expense["Date"])
            print("----------------------")

# Function to search expenses by category
def search_expense():
    search_category = input("Enter category to search: ").strip().title()
    found = False
    for expense in expenses:
        if expense["Category"] == search_category:
            print(expense)
            found = True
    if not found:
        print("No expenses found in this category.")

# Function to save expenses to file
def save_to_file():
    with open("expenses.txt", "w") as file:
        for expense in expenses:
            file.write(str(expense) + "\n")
    print("Expense records saved to file.")

# Main Menu
while True:
    print("\n===== Expense Tracker Menu =====")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Search Expense by Category")
    print("4. Save & Exit")

    choice = input("Enter your choice (1-4): ").strip()

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expenses()
    elif choice == "3":
        search_expense()
    elif choice == "4":
        save_to_file()
        print("Exiting program. Thank you!")
        break
    else:
        print("Invalid choice. Please select 1 to 4.")
```
